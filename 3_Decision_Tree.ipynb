**Data Pre-Processing**

```python
df_int = df.drop(obj_col, axis='columns')
df_int
```

```python
df_1 = df.drop(int_col, axis='columns')
df_1
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_encoded = df_1.apply(lb.fit_transform)
df_encoded
```

```python
df_cleaned = pd.concat([df_int, df_encoded], axis='columns')
df_cleaned
```

```python
x = df_cleaned.drop('deposit', axis='columns')
y = df_cleaned['deposit']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=3)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**Model Building**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini', max_depth=5, min_samples_split=10)
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
result = pd.DataFrame({
    'actual': y_test,
    'predicted': y_pred
})
result
```

**Model Evaluation**

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, classification_report
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

**Printing Decision Tree**

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(50,10))
plot_tree(dt, fontsize = 9, filled=True, rounded=True, feature_names=x.columns, class_names=['Deposited', 'Not Deposited']);
```
